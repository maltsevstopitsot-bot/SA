Завдання 5

Створіть два одновимірних масиви розміром 7, заповніть їх випадковими цілими числами від -15 до 15 та виконайте на них поелементні операції додавання, віднімання та множення.


```python
size_5 = 7
vector_a = np.random.randint(-15, 16, size=size_5)
vector_b = np.random.randint(-15, 16, size=size_5)

sum_5 = vector_a + vector_b
difference_5 = vector_a - vector_b
product_5 = vector_a * vector_b

print("\n5. Поелементні операції з векторами розміром 7 (цілі [-15, 15]):")
print(f"Вектор A: {vector_a}")
print(f"Вектор B: {vector_b}")
print(f"Сума (A + B): {sum_5}")
print(f"Різниця (A - B): {difference_5}")
print(f"Добуток (A * B): {product_5}")
```

Завдання 6

```python
matrix_a_6 = np.random.randint(1, 11, size=(4, 4))
matrix_b_6 = np.random.randint(1, 11, size=(4, 4))

product_6 = matrix_a_6 @ matrix_b_6

print("\n6. Матричний добуток матриць 4x4 (цілі [1, 10]):")
print("Матриця A:")
print(matrix_a_6)
print("Матриця B:")
print(matrix_b_6)
print("Добуток A @ B:")
print(product_6)
```

Завдання 8

Створіть матрицю розміром $2 \times 2$, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю.

```python
while True:
    matrix_8 = np.random.randint(1, 13, size=(2, 2))
    if np.linalg.det(matrix_8) != 0:
        break

inverse_matrix_8 = np.linalg.inv(matrix_8)

print("\n8. Обернена матриця 2x2 (цілі [1, 12]):")
print("Вихідна матриця:")
print(matrix_8)
print("Обернена матриця:")
print(inverse_matrix_8)
```

Завдання 15

Створіть матрицю розміром $5 \times 5$, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить середнє геометричне значення елементів кожного рядка матриці.

```python
matrix_15 = np.random.uniform(0.01, 0.51, size=(5, 5))

mean_geometric_row = gmean(matrix_15, axis=1)

vector_column_15 = mean_geometric_row.reshape(-1, 1)

print("\n15. Вектор-стовпчик середнього геометричного (Матриця 5x5, дійсні [0.01, 0.51]):")
print("Матриця:")
print(matrix_15)
print("Вектор-стовпчик середніх геометричних значень рядків (5x1):")
print(vector_column_15)
```
